```python
def ufun(x):
    return 1/(2-np.cos(x*np.pi))
ufun = np.vectorize(ufun)
```

```python
x,w = JacobiGQ(0,0,10)
```

```python
M = 200
N = [10,40,80,100,200]
#N = [10]
coefMat = np.empty((len(N),M))
alpha = 0
beta = 0

for i in range(len(N)):
    X, W = JacobiGQ(alpha=alpha,beta=beta,N=N[i])
    xx = np.diag(X)
    fvals = ufun(xx+1) #change x-axis
    P = JacobiP(xx,alpha=alpha,beta=beta,n=M-1,matrix=True)
    coef =  (P * fvals) @ W
    coefMat[i,] = coef
    plt.plot(coef,label=f"N={N[i]}")
plt.legend()
plt.xlabel("Coefficient number")
plt.show()

```
